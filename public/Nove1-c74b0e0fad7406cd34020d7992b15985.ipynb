# Kinematika pohybu vozidla

Tento notebook sa venuje základným pojmom kinematiky: pohybu s konštantnou rýchlosťou, rovnomerne zrýchlenému pohybu a brzdeniu vozidla.

Pomocou interaktívnych sliderov môžeš meniť hodnoty fyzikálnych veličín a sledovať, ako sa okamžite zmenia grafy závislostí polohy, rýchlosti a zrýchlenia od času.

---

## Fyzikálna teória

Pohyb vozidla opisujú základné rovnice kinematiky:

| Veličina | Vzorec | Jednotka |
|----------|--------|----------|
| Poloha | $s(t) = s_0 + v_0 t + \frac{1}{2} a t^2$ | m |
| Rýchlosť | $v(t) = v_0 + a \cdot t$ | m/s |
| Zrýchlenie | $a = \text{konšt.}$ | m/s² |

kde $v_0$ je počiatočná rýchlosť, $a$ je zrýchlenie a $t$ je čas.

> **Brzdná dráha:** Vzdialenosť ktorú vozidlo prejde od začiatku brzdenia po úplné zastavenie závisí od rýchlosti kvadraticky: $s = \frac{v_0^2}{2|a|}$. To znamená, že pri dvojnásobnej rýchlosti je brzdná dráha štvornásobná.

```python
# Import potrebných knižníc
import numpy as np                    # numerické výpočty
import matplotlib.pyplot as plt       # vykresľovanie grafov
import ipywidgets as widgets          # interaktívne ovládacie prvky
from IPython.display import display   # zobrazenie widgetov v notebooku

# Nastavenie tmavého štýlu grafov pre lepšiu čitateľnosť
plt.style.use('default')
```

## Interaktívna simulácia

Použi slidery na nastavenie počiatočnej rýchlosti a zrýchlenia.
Grafy sa aktualizujú okamžite.

> **Úloha 1:** Nastav zrýchlenie na zápornú hodnotu (brzdenie).
> Ako sa zmení tvar grafu polohy?

> **Úloha 2:** Zdvojnásob počiatočnú rýchlosť.
> O koľko sa zmenila brzdná dráha?

```python
def simulacia_kinematiky(v0, a):
    """
    Funkcia ktorá vypočíta a zobrazí grafy kinematiky pohybu vozidla.

    Parametre:
    v0 - počiatočná rýchlosť v m/s
    a  - zrýchlenie v m/s² (záporná hodnota = brzdenie)
    """

    # Vypočítame čas zastavenia ak vozidlo brzdí
    # Vozidlo sa zastaví keď v(t) = 0, teda t = -v0/a
    if a < 0 and v0 > 0:
        t_stop = -v0 / a    # čas zastavenia v sekundách
    else:
        t_stop = 10         # predvolený čas simulácie ak vozidlo nezastavuje

    # Vytvoríme pole 500 časových hodnôt od 0 po čas zastavenia
    # Viac hodnôt = plynulejšie grafy
    t = np.linspace(0, t_stop, 500)

    # Výpočet fyzikálnych veličín pre každý časový okamih
    s = v0 * t + 0.5 * a * t**2    # poloha v metroch podľa kinematickej rovnice
    v = v0 + a * t                  # rýchlosť v m/s
    acc = np.full_like(t, a)        # zrýchlenie je konštantné po celú dobu

    # Výpočet brzdnej dráhy ak vozidlo brzdí
    if a < 0 and v0 > 0:
        s_brzd = v0**2 / (2 * abs(a))   # vzorec pre brzdnú dráhu
    else:
        s_brzd = None

    # Vytvoríme tri grafy vedľa seba v jednom riadku
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))
    fig.suptitle(
        f'Kinematika pohybu vozidla  |  $v_0$ = {v0} m/s  |  $a$ = {a} m/s²',
        fontsize=13
    )

    # --- Graf 1: Poloha od času ---
    ax1.plot(t, s, color='#00BFFF', linewidth=2)
    ax1.set_xlabel('Čas (s)')
    ax1.set_ylabel('Poloha (m)')
    ax1.set_title('Poloha s(t)')
    ax1.grid(True, alpha=0.3)
    # Ak vozidlo brzdí, vyznačíme brzdnú dráhu červenou čiarkou
    if s_brzd is not None:
        ax1.axhline(y=s_brzd, color='red', linestyle='--', alpha=0.7,
                    label=f'Brzdná dráha: {s_brzd:.1f} m')
        ax1.legend()

    # --- Graf 2: Rýchlosť od času ---
    ax2.plot(t, v, color='#FFD700', linewidth=2)
    ax2.set_xlabel('Čas (s)')
    ax2.set_ylabel('Rýchlosť (m/s)')
    ax2.set_title('Rýchlosť v(t)')
    ax2.axhline(y=0, color='white', linestyle='-', alpha=0.3)  # nulová línia
    ax2.grid(True, alpha=0.3)

    # --- Graf 3: Zrýchlenie od času ---
    ax3.plot(t, acc, color='#FF6B6B', linewidth=2)
    ax3.set_xlabel('Čas (s)')
    ax3.set_ylabel('Zrýchlenie (m/s²)')
    ax3.set_title('Zrýchlenie a(t)')
    ax3.axhline(y=0, color='white', linestyle='-', alpha=0.3)  # nulová línia
    ax3.grid(True, alpha=0.3)

    plt.tight_layout()  # automatické rozmiestnenie grafov aby sa neprekrývali
    plt.show()

    # Textový súhrn výsledkov pod grafmi
    print(f'Počiatočná rýchlosť: {v0} m/s  ({v0*3.6:.1f} km/h)')
    if s_brzd is not None:
        print(f'Brzdná dráha: {s_brzd:.2f} m')
        print(f'Čas zastavenia: {t_stop:.2f} s')

# Vytvoríme interaktívne slidery prepojené so simulačnou funkciou
widgets.interact(
    simulacia_kinematiky,
    v0=widgets.FloatSlider(
        value=20,           # predvolená počiatočná rýchlosť 20 m/s
        min=0, max=50,      # rozsah 0 až 50 m/s
        step=1,
        description='v₀ (m/s):',
        style={'description_width': 'initial'}
    ),
    a=widgets.FloatSlider(
        value=-5,           # predvolené brzdenie -5 m/s²
        min=-20, max=10,    # záporné hodnoty = brzdenie, kladné = zrýchlenie
        step=0.5,
        description='a (m/s²):',
        style={'description_width': 'initial'}
    )
)
```

```python
# ÚLOHA 1: Zadanie číselnej odpovede s automatickou kontrolou
# Študent vypočíta brzdnú dráhu a zadá výsledok do textového poľa
# Systém odpoveď skontroluje a poskytne spätnú väzbu

print("📝 Otestuj svoje vedomosti:")
print("Pri počiatočnej rýchlosti 20 m/s a brzdení -5 m/s²")
print("aká je brzdná dráha? (výsledok zaokrúhli na celé číslo)")

# Textové pole kde študent zadá svoju odpoveď
odpoved_pole = widgets.Text(
    placeholder='Zadaj odpoveď v metroch...',
    description='Brzdná dráha:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='350px')
)

# Tlačidlo ktoré spustí kontrolu odpovede
tlacidlo = widgets.Button(
    description='Skontrolovať odpoveď',
    button_style='info',
    layout=widgets.Layout(width='200px')
)

# Výstupná oblasť kde sa zobrazí výsledok kontroly
vystup = widgets.Output()

def skontroluj_odpoved(b):
    """
    Funkcia ktorá skontroluje odpoveď študenta.
    Správna odpoveď: s = v0² / (2*|a|) = 20² / (2*5) = 400/10 = 40 m
    Povolená odchýlka je ±1 meter kvôli zaokrúhľovaniu.
    """
    with vystup:
        vystup.clear_output()   # vymaže predchádzajúci výstup pred novým
        try:
            # Načítame číselnú hodnotu zo vstupného poľa
            student = float(odpoved_pole.value)

            # Porovnáme so správnou odpoveďou s toleranciou ±1 m
            if abs(student - 40.0) <= 1:
                print('✅ Správne! Brzdná dráha je 40 m.')
                print('   Výpočet: s = v₀² / (2·|a|) = 400 / 10 = 40 m')
            else:
                print('❌ Skús znova. Použi vzorec: s = v₀² / (2·|a|)')
        except ValueError:
            # Ak študent nezadal číslo, zobrazíme upozornenie
            print('⚠️ Zadaj číselnú hodnotu.')

# Prepojenie tlačidla s kontrolnou funkciou
tlacidlo.on_click(skontroluj_odpoved)

# Zobrazenie všetkých prvkov v správnom poradí
display(odpoved_pole, tlacidlo, vystup)
```

```python
# ÚLOHA 2: Výber správnej odpovede z ponúknutých možností
# Študent vyberie jednu z možností a potvrdí výber tlačidlom
# Správna odpoveď je podložená fyzikálnym zdôvodnením

print("🔘 Vyber správnu odpoveď:")
print("Vodič ide rýchlosťou 60 km/h. Zdvojnásobí rýchlosť na 120 km/h.")
print("Ako sa zmení brzdná dráha?")

# Skupina prepínačov s možnosťami odpovede
# Len jedna možnosť môže byť vybratá naraz
moznosti = widgets.RadioButtons(
    options=[
        'Zdvojnásobí sa',
        'Strojnásobí sa',
        'Štvornásobí sa',
        'Zostane rovnaká'
    ],
    description='Odpoveď:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

# Tlačidlo na potvrdenie vybranej možnosti
tlacidlo_mc = widgets.Button(
    description='Potvrdiť odpoveď',
    button_style='info',
    layout=widgets.Layout(width='200px')
)

# Výstupná oblasť pre spätnú väzbu
vystup_mc = widgets.Output()

def skontroluj_mc(b):
    """
    Funkcia ktorá skontroluje vybranú možnosť.
    Správna odpoveď: Štvornásobí sa.
    Dôvod: s = v₀²/(2|a|) závisí kvadraticky od rýchlosti.
    Pri dvojnásobnej rýchlosti: s = (2v₀)²/(2|a|) = 4·v₀²/(2|a|)
    """
    with vystup_mc:
        vystup_mc.clear_output()    # vymaže predchádzajúci výstup

        if moznosti.value == 'Štvornásobí sa':
            # Správna odpoveď s fyzikálnym vysvetlením
            print('✅ Správne! Brzdná dráha závisí kvadraticky od rýchlosti.')
            print('   s = v₀² / (2·|a|)  →  pri dvojnásobnej rýchlosti je s štvornásobná.')
        else:
            # Nesprávna odpoveď s nápoveďou
            print('❌ Nie je správne. Nápoveda: pozri rovnicu s = v₀² / (2·|a|)')
            print('   Zamysli sa čo sa stane ak v₀ nahradíš hodnotou 2·v₀')

# Prepojenie tlačidla s kontrolnou funkciou
tlacidlo_mc.on_click(skontroluj_mc)

# Zobrazenie všetkých prvkov
display(moznosti, tlacidlo_mc, vystup_mc)
```

```python

#ÚLOHA 3: Doplňovačka - dosadzovanie do fyzikálneho vzorca
# Študent vypočíta tri hodnoty a dosadí ich do rovnice
# Každé pole sa kontroluje samostatne s individuálnou spätnou väzbou

print("✏️ Doplň správne hodnoty do rovnice brzdnej dráhy:")
print("Vozidlo ide rýchlosťou v₀ = 30 m/s a brzdí zrýchlením a = -6 m/s²")
print()
print("Brzdná dráha:  s = v₀² / (2 · |a|)  =  ___ / (2 · ___)  =  ___ m")
print()

# Vstupné pole pre prvú hodnotu: v₀² = 30² = 900
pole1 = widgets.FloatText(
    description='v₀² =',
    placeholder='?',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

# Vstupné pole pre druhú hodnotu: |a| = 6
pole2 = widgets.FloatText(
    description='|a| =',
    placeholder='?',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

# Vstupné pole pre výsledok: s = 900/12 = 75 m
pole3 = widgets.FloatText(
    description='s =',
    placeholder='? m',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='200px')
)

# Tlačidlo na overenie všetkých troch hodnôt naraz
tlacidlo_d = widgets.Button(
    description='Skontrolovať',
    button_style='info',
    layout=widgets.Layout(width='200px')
)

# Výstupná oblasť pre spätnú väzbu ku každému poľu
vystup_d = widgets.Output()

def skontroluj_doplnk(b):
    """
    Funkcia ktorá skontroluje všetky tri zadané hodnoty.
    Správne hodnoty:
    - v₀² = 30² = 900
    - |a|  = |-6| = 6
    - s    = 900 / (2·6) = 900 / 12 = 75 m
    """
    with vystup_d:
        vystup_d.clear_output()     # vymaže predchádzajúci výstup

        # Zoznam chýb ktoré sa zobrazia na konci
        chyby = []

        # Kontrola prvého poľa: v₀² musí byť 900
        if abs(pole1.value - 900) > 1:
            chyby.append('❌ v₀² nie je správne. v₀ = 30 m/s, teda v₀² = 30² = ?')
        else:
            print('✅ v₀² = 900 – správne!')

        # Kontrola druhého poľa: |a| musí byť 6
        if abs(pole2.value - 6) > 0.1:
            chyby.append('❌ |a| nie je správne. Zrýchlenie je -6 m/s², teda |a| = ?')
        else:
            print('✅ |a| = 6 – správne!')

        # Kontrola tretieho poľa: výsledok musí byť 75 m
        if abs(pole3.value - 75) > 1:
            chyby.append('❌ Výsledok nie je správny. Dosaď správne hodnoty do vzorca.')
        else:
            print('✅ s = 75 m – správne!')

        # Vypíšeme všetky chyby ak nejaké sú
        for chyba in chyby:
            print(chyba)

        # Ak nie sú žiadne chyby zobrazíme gratulačnú správu
        if not chyby:
            print()
            print('🎉 Výborne! Všetky hodnoty sú správne.')
            print('   s = 900 / (2·6) = 900 / 12 = 75 m')

# Prepojenie tlačidla s kontrolnou funkciou
tlacidlo_d.on_click(skontroluj_doplnk)

# Zobrazenie všetkých vstupných polí a tlačidla
display(pole1, pole2, pole3, tlacidlo_d, vystup_d)
```
