## <span style="color:#FFA500">Rovnomerne zrýchlený pohyb</span>

Pohyb opisujú tri základné rovnice kinematiky:

- **Poloha:** $s(t) = s_0 + v_0 \cdot t + \frac{1}{2} \cdot a \cdot t^2$
- **Rýchlosť:** $v(t) = v_0 + a \cdot t$
- **Zrýchlenie:** $a(t) = a = \text{konst.}$

kde $v_0$ je počiatočná rýchlosť a $a$ je zrýchlenie (záporné pri brzdení).

```python
# Import knižníc potrebných pre výpočty a vizualizáciu
import numpy as np                          # knižnica pre numerické výpočty
import matplotlib.pyplot as plt             # knižnica pre vykresľovanie grafov
import ipywidgets as widgets                # knižnica pre interaktívne ovládacie prvky
from IPython.display import display         # funkcia pre zobrazenie výstupov

%matplotlib inline                         
plt.style.use('dark_background')            # tmavý štýl pre všetky grafy
```

## <span style="color:#FFA500">Interaktívna simulácia pohybu</span>

Pomocou sliderov nastav:
- **v₀** – počiatočnú rýchlosť v km/h
- **a** – zrýchlenie (kladné) alebo spomalenie (záporné) v m/s²
- **Čas** – celkovú dĺžku simulácie v sekundách
- **Jednotky** – prepínač medzi metrami a kilometrami

> 💡 **Otázka 1:** Nastav v₀ = 50 km/h a a = −5 m/s². Za koľko sekúnd sa vozidlo zastaví?

> 💡 **Otázka 2:** Prečo je graf polohy zakrivený a graf rýchlosti priamka?

```python
# Hlavná funkcia pre výpočet a zobrazenie kinematiky
# Táto funkcia sa automaticky zavolá pri každej zmene slidera
def zobraz_kinematiku(v0_kmh, a, t_max, jednotky):

    # Prepočet počiatočnej rýchlosti z km/h na m/s
    v0 = v0_kmh / 3.6

    # Časová os – 500 rovnomerne rozmiestnených bodov
    t = np.linspace(0, t_max, 500)

    # Výpočet fyzikálnych veličín pomocou kinematických rovníc
    s     = v0 * t + 0.5 * a * t**2   # poloha [m]
    v     = v0 + a * t                  # rýchlosť [m/s]
    v_kmh = v * 3.6                     # rýchlosť prepočítaná na km/h
    acc   = np.full_like(t, a)          # zrýchlenie je konštantné [m/s²]

    # Orezanie na kladné hodnoty rýchlosti – vozidlo sa zastaví
    mask  = v >= 0
    t     = t[mask]
    s     = s[mask]
    v_kmh = v_kmh[mask]
    acc   = acc[mask]

    # Prepočet jednotiek polohy podľa výberu používateľa
    if jednotky == 'km':
        s_plot  = s / 1000
        s_label = 'Poloha s (km)'
    else:
        s_plot  = s
        s_label = 'Poloha s (m)'

    # Vytvorenie troch prepojených grafov so spoločnou osou x
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)
    fig.suptitle(
        f'Kinematika pohybu  │  v₀ = {v0_kmh} km/h  │  a = {a} m/s²',
        fontsize=13, color='white'
    )

    # Graf polohy
    ax1.plot(t, s_plot, color='#00BFFF', linewidth=2)
    ax1.set_ylabel(s_label, color='white')
    ax1.set_title('Závislosť polohy od času', color='#aaaaaa', fontsize=10)
    ax1.grid(True, alpha=0.2)
    ax1.fill_between(t, s_plot, alpha=0.1, color='#00BFFF')

    # Graf rýchlosti
    ax2.plot(t, v_kmh, color='#FFA500', linewidth=2)
    ax2.set_ylabel('Rýchlosť v (km/h)', color='white')
    ax2.set_title('Závislosť rýchlosti od času', color='#aaaaaa', fontsize=10)
    ax2.grid(True, alpha=0.2)
    ax2.fill_between(t, v_kmh, alpha=0.1, color='#FFA500')

    # Graf zrýchlenia
    ax3.plot(t, acc, color='#90EE90', linewidth=2)
    ax3.set_ylabel('Zrýchlenie a (m/s²)', color='white')
    ax3.set_xlabel('Čas t (s)', color='white')
    ax3.set_title('Závislosť zrýchlenia od času', color='#aaaaaa', fontsize=10)
    ax3.grid(True, alpha=0.2)
    ax3.fill_between(t, acc, alpha=0.1, color='#90EE90')

    plt.tight_layout()
    plt.show()

# Interaktívne slidery prepojené s funkciou
widgets.interact(
    zobraz_kinematiku,
    v0_kmh=widgets.FloatSlider(
        value=50, min=0, max=130, step=5,
        description='v₀ (km/h):',
        style={'description_width': '100px'},
        layout=widgets.Layout(width='500px')
    ),
    a=widgets.FloatSlider(
        value=-3, min=-10, max=5, step=0.5,
        description='a (m/s²):',
        style={'description_width': '100px'},
        layout=widgets.Layout(width='500px')
    ),
    t_max=widgets.FloatSlider(
        value=10, min=2, max=30, step=1,
        description='Čas (s):',
        style={'description_width': '100px'},
        layout=widgets.Layout(width='500px')
    ),
    jednotky=widgets.RadioButtons(
        options=['m', 'km'],
        value='m',
        description='Jednotky:',
        style={'description_width': '100px'}
    )
)
```

## <span style="color:#FFA500">Brzdná dráha</span>

Brzdná dráha závisí od počiatočnej rýchlosti **kvadraticky**. 
Pri dvojnásobnej rýchlosti je brzdná dráha **štvornásobná**.

$$s_{brzdenie} = \frac{v_0^2}{2 \cdot |a|}$$

> 💡 **Otázka 3:** Aká je brzdná dráha pri 50 km/h a aká pri 100 km/h? 
> Je druhá hodnota skutočne štvornásobná?

> 💡 **Otázka 4:** Ako sa zmení brzdná dráha ak zmeníš spomalenie z 6 m/s² na 3 m/s²?

```python
# Funkcia pre výpočet a zobrazenie brzdnej dráhy
def zobraz_brzdnu_drahu(a_brzd):

    # Rozsah rýchlostí od 0 do 130 km/h
    v_kmh  = np.linspace(0, 130, 200)
    v_ms   = v_kmh / 3.6                          # prepočet na m/s

    # Výpočet brzdnej dráhy podľa vzorca s = v₀²/(2|a|)
    s_brzd = v_ms**2 / (2 * abs(a_brzd))

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(v_kmh, s_brzd, color='#FF6B6B', linewidth=2.5)
    ax.fill_between(v_kmh, s_brzd, alpha=0.15, color='#FF6B6B')

    # Referenčné body pre bežné cestné rýchlosti
    for v_ref in [50, 90, 130]:
        s_ref = (v_ref / 3.6)**2 / (2 * abs(a_brzd))
        ax.plot(v_ref, s_ref, 'o', color='white', markersize=8)
        ax.annotate(
            f'{v_ref} km/h\n→ {s_ref:.1f} m',
            xy=(v_ref, s_ref),
            xytext=(v_ref + 4, s_ref + 3),
            fontsize=9, color='white'
        )

    ax.set_xlabel('Počiatočná rýchlosť (km/h)', color='white')
    ax.set_ylabel('Brzdná dráha (m)', color='white')
    ax.set_title(
        f'Brzdná dráha v závislosti od rýchlosti  │  spomalenie = {a_brzd} m/s²',
        color='white'
    )
    ax.grid(True, alpha=0.2)
    plt.tight_layout()
    plt.show()

# Slider pre brzdné spomalenie
widgets.interact(
    zobraz_brzdnu_drahu,
    a_brzd=widgets.FloatSlider(
        value=6, min=2, max=12, step=0.5,
        description='Spomalenie (m/s²):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px')
    )
)
```

## <span style="color:#FFA500">Zastavovacia vzdialenosť</span>

Celková zastavovacia vzdialenosť sa skladá z dvoch častí:

1. **Reakčná vzdialenosť** – vzdialenosť prejdená počas reakčného času
2. **Brzdná dráha** – vzdialenosť prejdená počas samotného brzdenia

$$s_{celková} = v_0 \cdot t_{reakcia} + \frac{v_0^2}{2 \cdot |a|}$$

> 💡 **Otázka 5:** Nastav rýchlosť 130 km/h a reakčný čas 1.5 s. 
> Porovnaj s reakčným časom 0.8 s. O koľko metrov je rozdiel?

> 💡 **Otázka 6:** Prečo je reakčná vzdialenosť lineárna ale brzdná dráha kvadratická?

```python
# Funkcia pre výpočet celkovej zastavovacej vzdialenosti
def vypocet_zastavovacia(v0_kmh, t_reakcia, a_brzd, jednotky):

    # Prepočet rýchlosti z km/h na m/s
    v0_ms = v0_kmh / 3.6

    # Reakčná vzdialenosť – vozidlo ide plnou rýchlosťou počas reakčného času
    s_reakcia  = v0_ms * t_reakcia

    # Brzdná dráha – kvadratická závislosť od rýchlosti
    s_brzdenie = v0_ms**2 / (2 * abs(a_brzd))

    # Celková zastavovacia vzdialenosť
    s_celkova  = s_reakcia + s_brzdenie

    # Prepočet jednotiek podľa výberu
    if jednotky == 'km':
        s_r, s_b, s_c, j = s_reakcia/1000, s_brzdenie/1000, s_celkova/1000, 'km'
    else:
        s_r, s_b, s_c, j = s_reakcia, s_brzdenie, s_celkova, 'm'

    # Horizontálny pruhový graf – oranžová = reakčná, červená = brzdná
    fig, ax = plt.subplots(figsize=(10, 3))
    ax.barh(['Vzdialenosť'], [s_r], color='#FFA500',
            label=f'Reakčná vzdialenosť: {s_r:.2f} {j}', height=0.4)
    ax.barh(['Vzdialenosť'], [s_b], left=[s_r], color='#FF6B6B',
            label=f'Brzdná dráha: {s_b:.2f} {j}', height=0.4)

    ax.set_xlabel(f'Vzdialenosť ({j})', color='white')
    ax.set_title(
        f'Zastavovacia vzdialenosť pri {v0_kmh} km/h  │  Celková: {s_c:.2f} {j}',
        color='white', fontsize=12
    )
    ax.legend(loc='lower right', fontsize=9)
    ax.grid(True, alpha=0.2, axis='x')
    plt.tight_layout()
    plt.show()

    # Číselný výpis výsledkov
    print(f'{"═"*45}')
    print(f'  Rýchlosť:              {v0_kmh} km/h')
    print(f'  Reakčný čas:           {t_reakcia} s')
    print(f'  Brzdné spomalenie:     {a_brzd} m/s²')
    print(f'{"─"*45}')
    print(f'  Reakčná vzdialenosť:   {s_r:.2f} {j}')
    print(f'  Brzdná dráha:          {s_b:.2f} {j}')
    print(f'{"─"*45}')
    print(f'  CELKOVÁ vzdialenosť:   {s_c:.2f} {j}')
    print(f'{"═"*45}')

# Slidery pre všetky parametre zastavovacej vzdialenosti
widgets.interact(
    vypocet_zastavovacia,
    v0_kmh=widgets.FloatSlider(
        value=90, min=10, max=200, step=5,
        description='Rýchlosť (km/h):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px')
    ),
    t_reakcia=widgets.FloatSlider(
        value=1.0, min=0.5, max=3.0, step=0.1,
        description='Reakčný čas (s):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px')
    ),
    a_brzd=widgets.FloatSlider(
        value=6, min=2, max=12, step=0.5,
        description='Spomalenie (m/s²):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px')
    ),
    jednotky=widgets.RadioButtons(
        options=['m', 'km'],
        value='m',
        description='Jednotky:',
        style={'description_width': '100px'}
    )
)
```
